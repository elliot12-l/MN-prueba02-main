## Grafique

````python
import matplotlib.pyplot as plt
# Ejecutar el método de Jacobi
A = [[1, 1], [-2, 5]]
b = [7, 0]
n = len(A)
x0 = [0]*n
# Obtener la solución y la trayectoria
solucion, trayectoria = gauss_jacobi(A=A, b=b, x0=x0, tol=1e-5, max_iter=100)
# Graficar
plt.figure(figsize=(10, 6))
plt.plot(trayectoria[:, 0], trayectoria[:, 1], 'bo-', label='Trayectoria de Jacobi')
plt.plot(x0[0], x0[1], 'ro', label='Estimación Inicial')
plt.plot(solucion[0], solucion[1], 'go', label='Punto de Solución')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Método de Jacobi: Trayectoria de la Solución')
plt.legend()
plt.grid(True)
plt.show()
````

# Resuelva

$ x_1 + x_2 = 6$

$  -2 x_1 +  x_2 = 0$

```
Gauss-Seidel no puede resolver la ecuacion dado que la diagonal de la matriz no es estrictamento dominante dado que la suma de los elemento de la segunda fila no cumple la condicion de dominancia puesto que 2>1, por endo se necesita usar otro metodo para la resolucion de la ecuacion
```

````python
import matplotlib.pyplot as plt
# Resolver el sistema y obtener la trayectoria
A = [[1, 1], [-2, 1]]
b = [6, 0]
n = len(b)
x0 = [0]*n
solucion, trayectoria = gauss_seidel(A=A, b=b, x0=x0, tol=1e-5, max_iter=10)
# Graficar
plt.figure(figsize=(10, 6))
plt.plot(trayectoria[:, 0], trayectoria[:, 1], 'bo-', label='Trayectoria de Gauss-Seidel')
plt.plot(x0[0], x0[1], 'ro', label='Estimación Inicial')
plt.plot(solucion[0], solucion[1], 'go', label='Punto de Solución')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Método de Gauss-Seidel: Trayectoria de la Solución')
plt.legend()
plt.grid(True)
plt.show()
````
